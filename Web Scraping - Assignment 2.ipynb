```python
import selenium
from selenium import webdriver
import pandas as pd
from selenium.webdriver.common.by import By
import warnings
warnings.filterwarnings("ignore")
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the shine.com page
driver.get("https://www.shine.com/")

time.sleep(2)

# Click on the search input field with "Data Scientist"
search_input = driver.find_element(By.XPATH, '//input[@class="input"]')
search_input.click()

time.sleep(2)

# Fill in the job title input field with "Data Scientist"
designation=driver.find_element(By.XPATH, "//input[@id='id_q']")
designation.send_keys("Data Analyst")

# Fill in the location input field
location = driver.find_element(By.XPATH, "//input[@id='id_loc']")
location.send_keys("Bangalore")

# Click the search button
search = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div[1]/form/div/div[2]/div/button")
search.click()

# Create an empty list to store job titles
job_title = []

# Find job title elements
title_tags = driver.find_elements(By.XPATH, '//a[contains(text(), "Data Analyst")]')

time.sleep(2)

# Loop through the job title elements and extract text for the first 10 jobs
for i in title_tags[:10]:
    title = i.text
    job_title.append(title)

# Create a DataFrame
df = pd.DataFrame({'Title': job_title})

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the shine.com page
driver.get("https://www.shine.com/")

time.sleep(2)

# Click on the search input field with "Data Scientist"
search_input = driver.find_element(By.XPATH, '//input[@class="input"]')
search_input.click()

time.sleep(2)

# Fill in the job title input field with "Data Scientist"
designation=driver.find_element(By.XPATH, "//input[@id='id_q']")
designation.send_keys("Data Scientist")

# Fill in the location input field
location = driver.find_element(By.XPATH, "//input[@id='id_loc']")
location.send_keys("Bangalore")

# Click the search button
search = driver.find_element(By.XPATH, "/html/body/div/div[4]/div/div[2]/div[2]/div[1]/form/div/div[2]/div/button")
search.click()

# Create empty lists to store job details
job_title = []
job_location = []
company_name = []

# Wait for job listings to load
time.sleep(5)

while len(job_title) < 10:

    # Find job title elements
    title_tags = driver.find_elements(By.XPATH, '//h2[@itemprop="name"]/a')
    for i in title_tags:
        title = i.text
        job_title.append(title)

    # Find job location elements
    location_tags = driver.find_elements(By.XPATH, '//div[@class=" jobCard_jobCard_lists_item__YxRkV jobCard_locationIcon__zrWt2"]')
    for i in location_tags:
        location = i.text
        job_location.append(location)

    # Find company name elements
    company_tags = driver.find_elements(By.XPATH, '//div[@class="jobCard_jobCard_cName__mYnow"]/span')
    for i in company_tags:
        company = i.text
        company_name.append(company)
        
# Ensure all lists are of the same length
min_length = min(len(job_title), len(job_location), len(company_name))
job_title = job_title[:min_length]
job_location = job_location[:min_length]
company_name = company_name[:min_length]

# Set the display.max_colwidth option to truncate the "Quote Type" column
pd.set_option('display.width', 1000)  # Set the desired width

# Create a DataFrame
df = pd.DataFrame({'Title': job_title[:10], 'Location': job_location[:10], 'Company_name': company_name[:10]})

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```

# Q4

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the flipkart.com page (requires login)
driver.get("https://www.flipkart.com/search?q=sunglasses&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off")

# Create empty lists to store product details
product_brand=[]
product_description=[]
product_price=[]
product_discount=[]

while len(product_brand) < 100:
    
    # Find product brands
    brand_tags=driver.find_elements(By.XPATH, '//div[@class="_2WkVRV"]')
    for i in brand_tags:
        brands=i.text
        product_brand.append(brands)

    # Find product description
    description_tags=driver.find_elements(By.XPATH, '//a[@class="IRpwTa"]')
    for i in description_tags:
        description=i.text
        product_description.append(description)

    # Find product price
    price_tags=driver.find_elements(By.XPATH, '//div[@class="_30jeq3"]')
    for i in price_tags:
        price=i.text
        product_price.append(price)

    # Find product discount
    discount_tags=driver.find_elements(By.XPATH, '//div[contains(@class,"_3Ay6Sb")]/span')
    for i in discount_tags:
        discount=i.text
        product_discount.append(discount)
    
    if len(product_brand) >= 100:
        break  # Exit the loop once you have 100 reviews

time.sleep(2)    

# Create a DataFrame
df = pd.DataFrame({'Brand': product_brand[:100], 'ProductDescription': product_description[:100], 'Price': product_price[:100], "Discount": product_discount[:100] })

# Print the DataFrame without line breaks
print(df.to_string(index=False))

# Close the WebDriver when you're done
driver.quit()
```

# Q5

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the Flipkart page
driver.get("https://www.flipkart.com/apple-iphone-11-black-64-gb/product-reviews/itm4e5041ba101fd?pid=MOBFWQ6BXGJCEYNY&lid=LSTMOBFWQ6BXGJCEYNYZXSHRJ&marketplace=FLIPKART")

# Create empty lists to store product details
rating = []
review_summary = []
full_review = []

while len(rating) < 100:  # Modify the condition to scrape up to 100 reviews
    # Find product rating
    rating_tags = driver.find_elements(By.XPATH, '//div[@class="_3LWZlK _1BLPMq"]')
    for i in rating_tags:
        rating.append(i.text)

    # Find review summary
    summary_tags = driver.find_elements(By.XPATH, '//p[@class="_2-N8zT"]')
    for i in summary_tags:
        review_summary.append(i.text)

    # Find full review
    review_tags = driver.find_elements(By.XPATH, '//div[@class="t-ZTKy"]/div/div')
    for i in review_tags:
        full_review.append(i.text)

    if len(rating) >= 100:
        break  # Exit the loop once you have 100 reviews

    # Scroll down to load more reviews
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)

# Create a DataFrame
df = pd.DataFrame({'Rating': rating, 'Review Summary': review_summary, 'Full Review': full_review})

# Set pandas options to display all columns in the same row and truncate the "Full Review" column if it exceeds a certain width
pd.set_option('display.max_colwidth', 40)  # Set the desired width for the "Full Review" column

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the Amazon page
driver.get("https://www.amazon.in/")

# Fill in the search input field with "Laptop"
search_input = driver.find_element(By.ID, 'twotabsearchtextbox')
search_input.send_keys("Laptop")

# Click the search button
search_button = driver.find_element(By.ID, 'nav-search-submit-button')
search_button.click()

# Wait for a moment for the page to load
time.sleep(2)

# Click on the "Intel Core i7" filter checkbox
wait = WebDriverWait(driver, 10)
cpu_type = wait.until(EC.element_to_be_clickable((By.XPATH, '//span[@class="a-size-base a-color-base" and contains(., "Intel Core i7")]')))
cpu_type.click()

# Create empty lists to store product details
product_title = []
product_price = []
product_rating = []

while len(product_title) < 10:
    # Find product title
    title_tags = driver.find_elements(By.XPATH, '//span[@class="a-size-medium a-color-base a-text-normal"]')
    for i in title_tags:
        title = i.text.split(',')[0]  # Split the title by comma and take the part before the first comma
        product_title.append(title)

    # Find product price
    price_tags = driver.find_elements(By.XPATH, '//span[@class="a-price-whole"]')
    for i in price_tags:
        price = i.text
        product_price.append(price)

    # Find product rating (if available)
    rating_tags = driver.find_elements(By.XPATH, '//span[@class="a-icon-alt"]')
    for i in rating_tags:
        rating = i.get_attribute("textContent")
        product_rating.append(rating)

    if len(product_title) >= 10:
        break  # Exit the loop once you have 10 product details

# Create a DataFrame
df = pd.DataFrame({'Title': product_title[:10], 'Price': product_price[:10], 'Rating': product_rating[:10]})

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```

# Q8

```python
# Create a WebDriver instance
driver = webdriver.Chrome()

# Create empty lists to store product details
quote = []
author = []
quote_type = []

for page in range(1, 11):  # Loop through pages 1 to 10
    # Opening the AZ Quotes page for the current page number
    driver.get(f"https://www.azquotes.com/top_quotes.html?p={page}")

    # Wait for a moment for the page to load
    time.sleep(2)

    # Find quote
    quote_tags = driver.find_elements(By.XPATH, '//a[@class="title"]')
    for i in quote_tags:
        quotes = i.text
        quote.append(quotes)

    # Find author
    author_tags = driver.find_elements(By.XPATH, '//div[@class="author"]/a')
    for i in author_tags:
        authors = i.text
        author.append(authors)

    # Find type of quote
    type_tags = driver.find_elements(By.XPATH, '//div[@class="tags"]')
    for i in type_tags:
        type_of_quote = i.text
        quote_type.append(type_of_quote)
    
    # Scroll down to load more reviews
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(2)  
    
# Set the display.max_rows option to None to display all rows
pd.set_option('display.max_rows', None)

# Set the display.max_colwidth option to truncate the "Quote Type" column
pd.set_option('display.max_colwidth', 40)  # Set the desired width

# Create a DataFrame
df = pd.DataFrame({'Quote': quote, 'Author': author, 'Quote Type': quote_type})

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()

# Opening the Jagran Josh page
driver.get("https://www.jagranjosh.com/")

gk_option = driver.find_element(By.XPATH, '//a[@href="https://www.jagranjosh.com/general-knowledge"]')
gk_option.click()

# Click consent using WebDriverWait
consent = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[@class="fc-button fc-cta-consent fc-primary-button"]')))
consent.click()

# Wait for a short time to ensure the page loads after consent
time.sleep(2)

explore_button = driver.find_element(By.XPATH, '//a[text()="India GK"]')
explore_button.click()

minister_list = driver.find_element(By.XPATH, '//img[@title="List of all Prime Ministers of India (1947-2023)"]')
minister_list.click()

# Click consent using WebDriverWait
consent = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, '//button[@class="fc-button fc-cta-consent fc-primary-button"]')))
consent.click()

# Wait for a short time to ensure the page loads after consent
time.sleep(2)

# Find all 'p' elements within the page
p_tags = driver.find_elements(By.XPATH, '//div[count(p) > 100]')

sn = []
pm_name = []
born_dead = []
term_of_office = []
duration_of_tenure = []
remark = []

for i in range(0, len(p_tags), 6):
    try:
        # Access the specific 'p' tag within the loop
        sn_tags = p_tags[i].find_element(By.XPATH, "./p[1]")
        sn_text = sn_tags.text
        sn.append(sn_text)

        pm_name_tags = p_tags[i].find_element(By.XPATH, "./p[2]")
        pm = pm_name_tags.text
        pm_name.append(pm)

        born_dead_tags = p_tags[i].find_element(By.XPATH, "./p[3]")
        birth_death = born_dead_tags.text
        born_dead.append(birth_death)

        term_tags = p_tags[i].find_element(By.XPATH, "./p[4]")
        office_term = term_tags.text
        term_of_office.append(office_term)

        duration_tags = p_tags[i].find_element(By.XPATH, "./p[5]")
        tenure_duration = duration_tags.text
        duration_of_tenure.append(tenure_duration)

        remark_tags = p_tags[i].find_element(By.XPATH, "./p[6]")
        remarks = remark_tags.text
        remark.append(remarks)

        i += 1
    except:
        break


# Scroll down to load more reviews
time.sleep(2)

# Set the display.max_colwidth option to truncate the "Quote Type" column
pd.set_option('display.max_colwidth', 30) 

# Create a DataFrame
df = pd.DataFrame({'S/N': sn, ' PM Name': pm_name, 'Born/Dead': born_dead, 'Term of Office': term_of_office, 'Duration of Tenure': duration_of_tenure, 'Remarks': remark})

# Print DataFrame
print(df)

# Close the WebDriver
driver.quit()

```

```python
# Create a WebDriver instance 
driver = webdriver.Chrome()
wait = WebDriverWait(driver, 10)

# Opening the Motor1 page
driver.get("https://www.motor1.com/")

# Click consent using WebDriverWait
consent = wait.until(EC.element_to_be_clickable((By.XPATH, '//button[@id="onetrust-accept-btn-handler"]')))
consent.click()

# Search for '50 most expensive cars'
search = wait.until(EC.presence_of_element_located((By.XPATH, '//input[@class="m1-search-panel-input m1-search-form-text"]')))
search.send_keys('50 most expensive cars')

# Click the search button
submit = wait.until(EC.element_to_be_clickable((By.XPATH, '//form[@id="header_search_form"]/button[@type="submit"]')))
submit.click()

# Click on the link for '50 Most Expensive Cars In The World'
expensive_cars = wait.until(EC.element_to_be_clickable((By.XPATH, '//a[text()="50 Most Expensive Cars In The World"]')))
expensive_cars.click()

car_model = []
car_image = []
car_price = []

while len(car_model) < 50:
    # Find car model
    car_tags = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//h3[@class="subheader"]/span')))
    for i in car_tags:
        car = i.text.split(',')[0]  # Split the title by comma and take the part before the first comma
        car_model.append(car)

    # Find car image
    image_tags = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//picture[@class="lazyload"]/img')))
    for i in image_tags:
        image = i.get_attribute('src')
        car_image.append(image)

    # Find car price
    price_tags = wait.until(EC.presence_of_all_elements_located((By.XPATH, '//p/strong')))
    for i in price_tags:
        price = i.text
        car_price.append(price)

    if len(car_model) >= 50:
        break  # Exit the loop once you have 50 product details
        
    # Scroll down to load more results
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")

# Create a DataFrame
df = pd.DataFrame({'Model': car_model[:50], 'Image': car_image[:50], 'Price': car_price[:50]})

# Print the DataFrame
print(df)

# Close the WebDriver when you're done
driver.quit()
```
